```haskell
-- Ejercicio 5 - Expresiones Lambda
--Lab01- Juan Diego Gaitan Haskell
--16/03/2026

--Reescribe estas funciones usando lambda:

--a. incrementarLista xs = map suma xs

--Sin lambda

suma :: Int -> Int
suma x = x + 1

incrementarLista :: [Int] -> [Int]
incrementarLista = map suma

--Con Lambda

incrementarListaLambda :: [Int] -> [Int]
incrementarListaLambda = map (\x -> x + 1) 

--Con Lambda Profesional (Sección de Operador)

incrementarListaOptimizada :: [Int] -> [Int]
incrementarListaOptimizada = map (+ 1)

incrementarLista [5,4,-2,3]
--incrementarLista [5,4,-2,3] = [6,5,-1,4]

--b. filtrarPares xs = filter even xs

--Sin Lambda

esPar :: Int -> Bool
esPar x = ((x `mod` 2) == 0)

filtrarPares :: [Int] -> [Int]
filtrarPares xs = filter esPar xs

--Con Lambda
filtrarParesLambda :: [Int] -> [Int]
filtrarParesLambda = filter (\x -> even x ) 


--Con Lambda Profesional 
filtrarParesLambdaPro :: [Int] -> [Int]
filtrarParesLambdaPro = filter even
filtrarParesLambdaPro [2,3,4,5,6,7,8,9,10]

--filtrarParesLambdaPro [2,3,4,5,6,7,8,9,10] = [2,4,6,8,10]

--c. sumarTuplas xs = ...

--Sin Lambda 

sumaTupla :: (Int, Int) -> Int
sumaTupla (x,y) = x + y

sumarTuplas :: [(Int,Int)] -> [Int]
sumarTuplas xs = map sumaTupla xs

--Con Lambda 
sumarTuplasLambda :: [(Int,Int)] -> [Int]
sumarTuplasLambda = map (\(x,y) -> x + y) 

--Con Lambda Pro 

sumarTuplasPro :: [(Int,Int)] -> [Int]
sumarTuplasPro = map (uncurry (+))

sumarTuplasPro [(1,5),(6,7),(-6,-10)]
--sumarTuplas [(1,5),(6,7),(-6,-10)] = [6,13,-16] 




```
