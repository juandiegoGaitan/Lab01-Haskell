```python
--Pregunta: ¿Qué hace la expresión [1..10] y cómo se relaciona con la evaluación perezosa?
--Representa una lista que empieza en 1 e itera uno en uno hasta 10. Sin embargo Haskell solo compila y reserva 
--memoria a los numeros de la lista en cuanto se necesitan (Ahorra memoria)

--2.Define las siguientes funciones

--Devuelve el primer elemento de la lista usando patrones.
primerElemento :: [a] -> a
primerElemento [] = -1
primerElemento (x:_) = primerElemento [x]
primerElemento [6,7]

-- primerlElemento [6,7] = 6

--Calcula el producto de todos los elementos de la lista de forma recursiva.
productoLista :: [Int] -> Int
productoLista [] = 1
productoLista (x:xs) = x * productoLista xs
productoLista [1,2,3]

--productoLista [1,2,3] = 6

--Devuelve una lista que contiene solo los números mayores o iguales a 0.
eliminarNegativos :: [Int] -> [Int]
eliminarNegativos [] = []
eliminarNegativos (x:xs) 
    | x >= 0    = x : eliminarNegativos xs
    | otherwise = eliminarNegativos xs

eliminarNegativos [1,2,-3,4,5,-7,-100]

--eliminarNegativos[1,2,-3,4,5,-7,-100] = [1,2,4,5]

--Invierte una lista sin usar reverse
invertir :: [a] -> [a]
invertir [] = []
invertir (x:xs) = invertir xs ++ [x]

invertir [1,2,-3,4,5,-7] 

--invertir [1,2,-3,4,5,-7]  = [-7,5,4,-3,2,1]


```
