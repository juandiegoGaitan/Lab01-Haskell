```haskell
-- Ejercicio 03 Definición de funciones: composición y casos- 
--Lab01- Juan Diego Gaitan Haskell
--16/03/2026

--1. Defina las funciones: 

--a. Calcula la suma de los cuadrados de dos número
suma :: Int -> Int -> Int 
suma x y = x + y

cuadrado :: Int -> Int 
cuadrado x = x*x

sumaCuadrados :: Int -> Int -> Int 
sumaCuadrados x y = suma (cuadrado x) (cuadrado y)

sumaCuadrados 1 2 

--sumaCuadrados 1 2 = 5

--b. Multiplica por 2 cada elemento de una lista usando map y una funcion "duplicar".

duplicar :: Int -> Int
duplicar x = x*2

dobleLista :: [Int] -> [Int]
dobleLista = map duplicar 

dobleLista [1,5,2]

--Forma más elegante

dobleListaSimple :: [Int] -> [Int]
dobleListaSimple = map (*2)

dobleLista [1,5,2]
--dobleListaSimple [1,5,2] = [2,10,4]


--2. Define las siguientes funciones:

--a. Implementar clasificarNota según :

 --Menor a 3 → "Reprobado"
  
  --Entre 3 y 4 → "En recuperación"
  
  --Mayor o igual a 4 → "Aprobado"

clasificarNota :: Int -> String
clasificarNota x 
    |x < 3     = "Reprobado" 
    |x <= 4    = "En recuperacion" 
    |otherwise = "Aprobado"
clasificarNota 6

--clasificarNota 6 = "Aprobado"


--b. Implementar fibonacci
fibonacci :: Int -> Int
fibonacci 0 = 0
fibonacci 1 = 1
fibonacci x = fibonacci (x-1) + fibonacci (x-2)

fibonacci 4

--fibonacci 2 = 1
```
